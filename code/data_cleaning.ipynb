```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../dataset/ny_house_data.csv')
df
```

```python
df.tail()
```

```python
df.shape
```

```python
selected_columns = ['TYPE', 'PRICE', 'BEDS', 'BATH', 'PROPERTYSQFT', 'STATE', 'STREET_NAME', 'LATITUDE', 'LONGITUDE']
df_selected = df[selected_columns]
df_selected.head()
```

```python
df_selected.columns = [x.lower() for x in df_selected.columns]
df_selected.head()
```

```python
df_selected = df_selected.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
df_selected
```

```python
df_selected['zip_code']= df_selected['state'].str.extract(r'(\d{5})$')
df_selected['city']= df_selected['state'].apply(lambda x: x.split(',')[0].strip())
df_selected
```

```python
df_selected = df_selected.drop(columns=['state'])
df_selected
```

```python
df_selected.dtypes
```

```python
#counting number of house by house type

number_of_house_by_type = df_selected['type'].value_counts()
number_of_house_by_type
```

```python
# creating a bar graph to visualize number of house by house type
number_of_house_by_type.plot(kind = 'bar')
plt.title('Number of houses for different house type')
plt.xlabel('House Type')
plt.ylabel('Number of House')
```

```python
#counting number of house by number of bedrooms

number_of_house_by_beds = df_selected['beds'].value_counts()
number_of_house_by_beds
```

```python
# creating a bar graph to visualize number of house by house type
number_of_house_by_beds.plot(kind = 'bar', figsize=(10, 6))
plt.title('Number of houses for different Bedroom Numbers')
plt.xlabel('Bedroom Number')
plt.ylabel('Number of House')
```

```python
#counting number of house by house City

number_of_house_by_city = df_selected['city'].value_counts()
number_of_house_by_city
```

```python
# creating a bar graph to visualize number of house by city
number_of_house_by_city.plot(kind = 'bar', figsize=(10, 6))
plt.title('Number of houses for different city')
plt.xlabel('City')
plt.ylabel('Number of House')
```

```python
#counting number of house by zipcode

number_of_house_by_zipcode = df_selected['zip_code'].value_counts()
number_of_house_by_zipcode
```

```python
# creating a bar graph to visualize number of house by zipcode
# number_of_house_by_zipcode.plot(kind = 'bar', figsize=(10, 6))
plt.figure(figsize=(10, 6)) 
plt.bar(number_of_house_by_zipcode.index[:50],number_of_house_by_zipcode[:50],color='green')
plt.xticks(rotation=90)
plt.title('Number of houses for top 50 Zipcode')
plt.xlabel('Zipcode')
plt.ylabel('Number of House')
```

```python
#counting number of house by street name (area name)

number_of_house_by_street = df_selected['street_name'].value_counts()
number_of_house_by_street
```

```python
# creating a bar graph to visualize number of house by area name

plt.figure(figsize=(10, 6)) 
plt.bar(number_of_house_by_street.index[:25],number_of_house_by_street[:25],color='purple')
plt.xticks(rotation=90)
plt.title('Number of houses for top 25 area name')
plt.xlabel('street / area name')
plt.ylabel('Number of House')
```
