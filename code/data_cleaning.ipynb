```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../dataset/ny_house_data.csv')
df
```

```python
df.tail()
```

```python
df.shape
```

```python
selected_columns = ['TYPE', 'PRICE', 'BEDS', 'BATH', 'PROPERTYSQFT', 'STATE', 'STREET_NAME', 'LATITUDE', 'LONGITUDE']
df_selected = df[selected_columns]
df_selected.head()
```

```python
df_selected.columns = [x.lower() for x in df_selected.columns]
df_selected.head()
```

```python
df_selected = df_selected.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
df_selected
```

```python
df_selected['zip_code']= df_selected['state'].str.extract(r'(\d{5})$')
df_selected['city']= df_selected['state'].apply(lambda x: x.split(',')[0].strip())
df_selected
```

```python
df_selected = df_selected.drop(columns=['state'])
df_selected
```

df_selected.to_csv('../dataset/selected_dataset.csv', index = False)

## Since we have our optimised dataset we will start ploting data for each variable (columns) to visualized the structure and spread of the data

## ----------------- Visualize data by House Type ---------------

```python
#counting number of house by house type

number_of_house_by_type = df_selected['type'].value_counts()
number_of_house_by_type
```

```python
# creating a bar graph to visualize number of house by house type
number_of_house_by_type.plot(kind = 'bar')
plt.title('Number of houses for different house type', y=1.02)
plt.xlabel('House Type')
plt.ylabel('Number of House')
```

## House type visualization: 
**Here we see most number of listed houses are 'co-op's. 2nd, 3rd & 4th highest listed house types are 'Private House' (house), 'Condo' & 'Multi-family' respectively. All other category of houses are very few compare to the top 4. From here can say Most Number of houses that are sold in NYC are spread in 4 main categories.**

```python
# Removing 'land for sale', 'for sale', 'foreclosure', 'condop for sale', 'coming soon', 'mobile house for sale' categories,
# Since they are very little in number & also confusing types

categories_to_remove = ['land for sale', 'for sale', 'foreclosure', 'condop for sale', 'coming soon', 'mobile house for sale']
df_selected = df_selected[~df_selected['type'].isin(categories_to_remove)]

df_selected['type'].value_counts()
```

## ------------------------- Visualize Data By Number of Bedrooms --------------------

```python
#counting number of house by number of bedrooms

number_of_house_by_beds = df_selected['beds'].value_counts()
number_of_house_by_beds
```

```python
# creating a bar graph to visualize number of house by house type
number_of_house_by_beds.plot(kind = 'bar', figsize=(10, 6))
plt.title('Number of houses for different Bedroom Numbers', y=1.02)
plt.xlabel('Bedroom Number')
plt.ylabel('Number of House')
```

## Number of bedrooms visualization:
**Here we see Highest Listed Houses has respectively 3, 2, 1 & 4 bedrooms. 3 bedroom houses seems to be the most popular ones may be because of the the convenient it gives when you want a full family house** 

**There are also some houses listed with 20 to 50 bed rooms and I believe this are whole buildings for sale**

```python
# we will remove houses that has more than 12 bed bedrooms because any thing above this number has very few houses
# & can easily be considered as exception

df_selected = df_selected[df_selected['beds']<= 12]
df_selected['beds'].value_counts()
```

## ----------------- Visualize data by Median Price of House Type ---------------
